```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
# using PCA for dimentionality reduction
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x=pca.fit_transform(x)
```

```python
plt.figure(figsize=(12,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

```python
# I have found that among all the 562 features only approx 240 features are givinf majority of variances of the dataset
```

```python
pca=PCA(n_components=240)
x=pca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
# model building with logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
param_grid = {'C':[1,10,20,50],'gamma':[1,0.1,0.001,0.0001], 'kernel':['linear','rbf']}
```

```python
grid_search=GridSearchCV(SVC(),param_grid=param_grid,cv=5)
```

```python
grid_search.fit(x_train,y_train)
grid_search.best_params_
```

```python
svc=SVC(C=50,gamma=0.001,kernel='rbf')
svc.fit(x_train,y_train)
pred_svc=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,roc_auc_score
```

```python
print(classification_report(y_test,pred_svc))
```

```python
# model building with knn
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
parameters = {'n_neighbors':[4,5,6,7,9],
              'leaf_size':[1,3,5],
              'algorithm':['auto', 'kd_tree'],
              'n_jobs':[-1],
               'weights': ['uniform','distance']}

#Fit the model
model = GridSearchCV(knn, param_grid=parameters,cv=5)
model.fit(x_train,y_train)

#predictions on test data
prediction=model.predict(x_test)
```

```python
model.best_params_
```

```python
knn=KNeighborsClassifier(leaf_size=1,n_jobs=-1,n_neighbors=6,weights='distance')
```

```python
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_test)
```

```python
print(classification_report(y_test,pred_knn))
```
